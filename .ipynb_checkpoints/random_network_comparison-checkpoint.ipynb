## Creating Networks

```python
# data processing and network processing 
import pandas as pd
import networkx as nx

# plotting
import matplotlib
import matplotlib.pyplot as plt

# role extraction
from graphrole import RecursiveFeatureExtractor, RoleExtractor

# extras
import numpy as np
import seaborn as sns
import warnings
from pprint import pprint
from scipy import spatial
```

```python
# load the data set
data = pd.read_csv('data-triell.csv')
```

```python
def create_directed_graph(df):
    g = nx.from_pandas_edgelist(df, source='username_1',
                                target='username_2',
                                create_using=nx.DiGraph())
    return g
```

```python
# create retweet graph from dataframe
g = create_directed_graph(data)
```

### Role Discovery on Random Network

```python
# first test out role extraction on a random network (probability set to 0.0006 to simulate a similar average degree)
h = nx.erdos_renyi_graph(4114, 0.0006, directed=True)
```

```python
# extract features using graphrole library
h_feature_extractor = RecursiveFeatureExtractor(h)
h_features = h_feature_extractor.extract_features()
```

```python
# assign node roles
h_role_extractor = RoleExtractor(n_roles=None)
h_role_extractor.extract_role_factors(h_features)
h_node_roles = h_role_extractor.roles

print('\nh_node_roles:')
pprint(h_node_roles)

print('\nh_nodepercentage:')
print(h_role_extractor.role_percentage.round(2))
```

```python
#print list of usernames with specific role and add to a list
# this function is used here to check how many nodes are assigned to each role

username0_list = []
for username, role in h_node_roles.items():
    if role == 'role_7': # user can change role here to check
        username0_list.append(username)
```
